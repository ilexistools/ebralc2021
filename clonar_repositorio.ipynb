Vamos testar o acesso às funções do repositório clonado:

```python
from html2freq.html2freq import html2freq
import nltk  

# instala o requisito punkt se não for achado  
try:
  nltk.data.find('tokenizers/punkt')
except LookupError:
  nltk.download('punkt')

# conta palavras de uma página da wikipédia 
url = 'https://pt.wikipedia.org/wiki/Tubar%C3%A3o'
[item for item in html2freq(url).most_common() if len(item[0]) > 3]
```
